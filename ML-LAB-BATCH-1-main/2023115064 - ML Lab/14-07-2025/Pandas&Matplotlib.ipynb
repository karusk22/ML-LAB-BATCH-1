# Implementation of Python Libraries for ML application such as Pandas and Matplotlib.

**a) Create a Series using pandas and display**

```python
import numpy as np
import pandas as pd

s = pd.Series([1, 3, 5, np.nan, 6, 8])
s
```

**b) Access the index and the values of our Series**

```python
print("Index of the Series:", s.index)
print("Values of the Series:", s.values)
```

**c) Compare an array using Numpy with a series using pandas**

```python
import numpy as np
arr = np.array([1, 3, 5, np.nan, 6, 8])
comparison = s == arr
print("Comparison of Series and Array:\n", comparison)
```

**d) Define Series objects with individual indices**

```python
import pandas as pd
data = {'a': 0., 'b': 1., 'c': 2.}
s_indexed = pd.Series(data)
print("Series with individual indices:\n", s_indexed)

s_indexed_explicit = pd.Series([1, 2, 3], index=['x', 'y', 'z'])
print("Series with explicit indices:\n", s_indexed_explicit)
```

**e) Access single value of a series**

```python
print("Value at index 0:", s.iloc[0])
print("Value at label 'a':", s_indexed['a'])
```
